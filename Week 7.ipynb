Advantages that might pop up with using normalized data are that there'll be lest drastic results appearing, and they'll be (likely) easier to graph. This itself can also be a disadvantage because where there are big differences, they'll look small. Another reason as to why normalizing data is a good idea is that if there are different ways of rating films, scoring whatever it is that you're scoring - eg: 1-5 vs 1-10 vs 1-100, normalization will allow us to take a look at all of them without skewing the graphs that we draw.
