Let's work on titanic first. A quick scroll through the file shows us a handful of NaNs in the age. Let's assume that makes the data useless to us. There's a column for the sex, so SexCode is irrelevant. We'll remove that column completely. Lastly, let's get rid of anyone and everyone that didn't survive. We're only interested in the survivors - we'll assume that survivors are 1, and those that passed away are 0; after, we'll drop that column. We'll take a glance at the top 25 to make sure it's giving us what we want, then get the average sage of survivors at the end. 

```python
titanic_copy = titanic.copy()
titanic_copy.loc[~(titanic_copy['Survived'] == 0)]
survived = titanic_copy.drop(['Survived'], axis=1)
dropsex = survived.drop(['SexCode'], axis=1)
titanic_clean = dropsex.dropna(subset = ['Age'])
titanic_clean.head(25)
```

```python
print (titanic_clean["Age"].mean())
```

The average age of survivors - male, female combined is 30.39 years of age, and the rows/columns look far cleaner now. Yay.

Let's move onto the security csv.
There's a very wide column called Web.Description - it makes the spreadsheet look unsightly. We're getting rid of that, along with the covered entity type. Most of these are health-related anyway. We'll do that in one line instead of multiple like how we did above. That's not pretty after all. Next, let's turn the business associate present booleans to 0s and 1s, because those values might be of more use to us in the future. Afterwards, let's order by the most number of individuals affected and get the top 10.

```python
security_copy = security.copy()
dropdesc = security_copy.drop(['Web.Description', 'Covered.Entity.Type'], axis=1)
dropdesc.rename(columns={'Business.Associate.Present': 'Associate'}, inplace=True)
dropdesc.Associate = dropdesc.Associate.astype(int)
dropassociate = dropdesc.loc[~(dropdesc['Associate'] == 0)]
dropsex = survived.drop(['SexCode'], axis=1)
droppedass = dropassociate.drop(['Associate'], axis=1)
securitydone = droppedass.sort_values(by='Individuals.Affected', ascending=False)
securitydone.head(10)
```

Next we've got to go after the Global Terrorism Database yearly summaries csv. We're going to get rid of all the rows that have 0 incidents, then sort by descending frequencies.

```python
incidents_copy = incidents.copy()
incidents_copy.loc[~(incidents_copy['Freq'] == 0)]
sortedincidents = incidents_copy.sort_values(by='Freq', ascending=False)
sortedincidents.head(10)
```
