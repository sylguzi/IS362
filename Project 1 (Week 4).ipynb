First we need to import pandas & the .csv file. I have modified the .csv to my liking

```python
import pandas as pd
import numpy as np
flights=pd.read_csv("C:/Users/sguzi/Desktop/flights3.csv")
flights
```

Now that we've imported the data, let's make a copy to safeguard the original. We'll be working on this one throughout the project.

```python
flights_copy = flights.copy()
```

Let's get the total number of flights for each Airline/City

```python
flights_copy['Total'] = flights_copy.sum(axis= 1)
flights_copy
```

Now let's get the percentage of delayed flights for each airline to each city

```python
flights_copy['Percent_Delay'] = flights_copy['Delayed'].div(flights_copy.Total, axis = 0)
flights_copy
```

Let's sort it by the smallest ratio of delays, so that we know what Airline and what city have the best management.

```python
flights_copy.sort_values(by='Percent_Delay')
```

Looks like the best airline option is ALASKA, with a 5.1% chance of getting delayed, and the best city is Phoenix, with both airlines coming in with less than 10% chance of a delay!
